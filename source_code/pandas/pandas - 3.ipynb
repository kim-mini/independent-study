# loc

```
import pandas as pd
import numpy as np
```

```
df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                  index=["a", "b", "c"],
                  columns=["A", "B", "C", "D"])
df
```

```
df.loc['a']
```

```
df.loc[['b','c']]
```

```
df.A > 15
```

```
df.loc[df.A > 15]
```

```
df2 = pd.DataFrame(np.arange(10,26).reshape(4,4), columns=['a','b','c','d'])
```

```
df2
```

```
df2.loc[1:2]
```

# 연습 문제 

  - 모든 행과 열에 라벨을 가지는 5 x 5 이상의 크기를 가지는 데이터프레임을 만든다.

  - 10가지 이상의 방법으로 특정한 행과 열을 선택한다.

```
df3 = pd.DataFrame(np.arange(1,26).reshape(5,5), columns=['A','B','C','D','E'])
df3
```

```
df3.loc[0]
```

```
df3.loc[2:4]
```

```
df3.iloc[2:4,2:4]
```

```
df3.loc[:,'A':'C']
```

```
df3.loc[4,'B']
```

```
df3.loc[df3.A < 10,['B','C']]
```

```
df3.loc[[2,4],['A','C']]
```

```
df3.loc[3:]
```

```
df3.iloc[:,2]
```

```
df3.iloc[2] = df3.iloc[2]*100
```

```
df3
```
